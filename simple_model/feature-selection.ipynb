```python
df = pd.read_parquet('../input/orvpnoleak/orvp-features.parquet')
```

```python
X_train = df[[x for x in list(df) if x not in ['time_id', 'target', 'row_id']]]
X_train.replace([np.inf, -np.inf], np.nan, inplace = True)
X_train = X_train.fillna(X_train.mean())
y_train = df['target']
```

#### Set n_trial to something in the range of 20-40 for optimal results, 12 was used here because of Kaggle's runtime limit.

```python
from BorutaShap import BorutaShap
from lightgbm import LGBMRegressor 

Feature_Selector = BorutaShap(model = LGBMRegressor(categorical_column = [0],
                                                    learning_rate = 0.05,
                                                    n_estimators=500,
                                                    min_data_in_leaf = 500), 
                              importance_measure='shap', 
                              classification=False)

Feature_Selector.fit(X=X_train, y=y_train, n_trials=12, random_state=0)
```

```python
Feature_Selector.plot(which_features='all', figsize=(100,50))
```
